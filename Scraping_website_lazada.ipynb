```python
class preprocess:
    def low(self,data):
        return data.lower()
    def del_num(self,data):
        data = re.sub(r"\d", "", data)
        return data
    def del_punct(self,data):
        data = re.sub('[^A-Za-z0-9]+', ' ', data)
        return data
    def del_ws(self,data):
        return data.strip()
    def del_stopword(self,data):
        stopwords = ['dan', 'yang', 'untuk','pada','ke', 'karena','oleh', 'akan', 'dengan']
        data = ' '.join([word for word in data.split() if not word in stopwords])
        return data
    def proses_1(self,data):
        data = self.low(data)
        data = self.del_num(data)
        data = self.del_punct(data)
        data = self.del_ws(data)
        data = self.del_stopword(data)
        return data     
```

```python
df['Nama'] = df['Nama'].apply(lambda x:preprocess().proses_1(x))
    #print(kolom)
df
```

```python
class preprocess_hrg:
    def low(self,data):
        return data.lower()
    def del_punct(self,data):
        data = re.sub('rp', ' ', data)
        return data
    def del_ws(self,data):
        return data.strip()
    def del_num(self,data):
        data = re.sub(r"\n", "", data)
        return data
    def proses_hrg(self,data):
        data = self.low(data)
        data = self.del_punct(data)
        data = self.del_ws(data)
        data = self.del_num(data)
        return data     
```

```python
df['Harga'] = df['Harga'].apply(lambda x:preprocess_hrg().proses_hrg(x))
    #print(kolom)
df
```
